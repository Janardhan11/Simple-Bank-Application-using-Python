```python
class User:
    def __init__(self):
        self.user = {'1234567': {'Name':'Janardhan', 'Balance':7000}}

    def login(self):
        self.account_no = input('Enter your account_no: ')
        if self.account_no not in self.user:
            print('User name doesnt exist or Incorrect user name')
            return self.new_registration()
        else:
            print('Login Successful!')
            print('Welcome', self.user[self.account_no]['Name'])

        print('--------------------------------------------------------------------')
        print('press 1 for check balance')
        print('press 2 for deposite')
        print('press 3 for withdraw')
        print('press 0 for exit')
        print('--------------------------------------------------------------------')
        choice = int(input('Enter your choice: '))
        if choice == 1:
          self.check_balance(self.account_no, self.user)
        elif choice == 2:
          self.add_money(self.account_no, self.user)
        elif choice == 3:
          self.withdraw(self.account_no, self.user)
        else:
          print('Thankyou for visiting us get back soon!')


    def new_registration(self):
        print('--------------------------------------------------------------------')
        self.account_no = input('Create your 7 digit account_no: ')
        if len(self.account_no) != 7:
          print('Account_no should be 7 digits!')
          return self.new_registration()
        if self.account_no in self.user:
            print('Account_no already exist!')
            print('--------------------------------------------------------------------')
            print('press 1 for login')
            print('press 2 for new_registration')
            print('press 0 for exit')
            print('--------------------------------------------------------------------')
            choice = input('Enter your choice: ')
            if choice == 1:
              self.login()
            elif choice == 2:
              self.new_registration()
            elif choice == 0:
              print('Thankyou for visiting us get back soon!')
        else:
            self.Name = input('Enter your name: ')
            self.Balance = int(input('How much balance you want to deposite (Min 500): '))
            if self.Balance < 500:
              print('Deposite amount must be greater than 500')
              return self.new_registration()
            else:
              self.user[self.account_no] = {'Name': self.Name, 'Balance': self.Balance }
              print('Your registration is successful Thankyou!')
        print('--------------------------------------------------------------------')
        print('press 1 for login into existing account')
        print('press 0 for exit')
        print('--------------------------------------------------------------------')
        choice = int(input('Enter your choice: '))
        if choice == 1:
          self.login()
        else:
          print('Thankyou for visiting us get back soon!')


    def check_balance(self, account_no, user):
        print('Balance for your account_no',account_no,":" ,self.user[account_no]['Balance'])
        print('--------------------------------------------------------------------')
        print('press 1 for deposite')
        print('press 2 for withdraw')
        print('press 0 for exit')
        print('--------------------------------------------------------------------')
        choice = int(input('Enter your choice: '))
        if choice == 1:
          self.add_money(self.account_no, self.user)
        elif choice == 2:
          self.withdraw(self.account_no, self.user)
        else:
          print('Thankyou for visiting us get back soon!')

    def add_money(self,account_no, user):
        m = int(input('Enter money how much you want to deposite: '))
        self.user[account_no]['Balance']= self.user[account_no]['Balance'] + m
        print('Balance for your account_no',account_no,":" ,self.user[account_no]['Balance'])
        print('--------------------------------------------------------------------')
        print('press 1 for deposite')
        print('press 2 for withdraw')
        print('press 0 for exit')
        print('--------------------------------------------------------------------')
        choice = int(input('Enter your choice: '))
        if choice == 1:
          self.add_money(self.account_no, self.user)
        elif choice == 2:
          self.withdraw(self.account_no, self.user)
        else:
          print('Thankyou for visiting us get back soon!')

    def withdraw(self,account_no, user):
        m = int(input('Enter money how much you want to withdraw: '))
        if (self.user[account_no]['Balance'] <= 500) or (self.user[account_no]['Balance'] < m):
          print('Low balance')
          print('Add balance into your account please.')
          return self.add_money(self.account_no, self.user)
        else:
          self.user[account_no]['Balance']= self.user[account_no]['Balance'] - m
        print('Balance for your account_no',account_no,":" ,self.user[account_no]['Balance'])
        print('--------------------------------------------------------------------')
        print('press 1 for Details of your account')
        print('press 2 for withdraw')
        print('press 3 for deposite')
        print('press 0 for exit')
        print('--------------------------------------------------------------------')
        choice = int(input('Enter your choice: '))
        if choice == 1:
          self.print_user_details()
        elif choice == 2:
          self.withdraw(self.account_no, self.user)
        elif choice == 3:
          self.add_money(self.account_no, self.user)
        else:
          print('Thankyou for visiting us get back soon!')

    def option(self):
        print('--------------------------------------------------------------------')
        print('press 1 for New_Registration')
        print('press 2 for login into existing account')
        print('press 0 for exit')
        print('--------------------------------------------------------------------')
        choice = int(input('Enter your choice: '))
        if choice == 1:
            self.new_registration()
        elif choice == 2:
            self.login()
        else:
            print('Invalid choice')


    def print_user_details(self):
        print('--------------------------------------------------------------------')
        print("User Details:")
        for account_no, user_details in self.user.items():
            print(f"Account Number: {account_no}")
            print(f"Name: {user_details['Name']}")
            print(f"Balance: {user_details['Balance']}")
            print('--------------------------------------------------------------------')
            print()

bank = User()
bank.option()
```
